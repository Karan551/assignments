```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Data Structure : [resturant_id,2021,2022,2023,2024]

sales_data=np.array([
    [1, 150000, 180000, 220000, 250000], # Paradise Biryani
    [2, 120000, 140000, 160000, 190000], # Bejing Bites
    [3, 200000, 230000, 260000, 300000], # Pizza Hub
    [4, 180000, 210000, 240000, 270000], # Burger Point
    [5, 160000, 185000, 205000, 230000]  # Chai Point
    ])
print("======Zomato Sales Analysis======")
print("Sales data shape::",sales_data.shape)
print("Sample data for 1st 3 resturant::\n",sales_data[:3])
```

- axis=0 ðŸ‘‰ column wise sum <br/>
- axis=1 ðŸ‘‰ row wise sum 
- Syntax for slicing of numpy array :-
    - `data[row,column]`

```python
# Total Sales Per year 
print(np.sum(sales_data,axis=0))

yearly_total_sales=np.sum(sales_data[:,1:],axis=0)
print("yearly total sales::\n",yearly_total_sales)

# minimum sales per year
min_sales=np.min(sales_data[:,1:],axis=0)
# print("minimum sales per year::",min_sales)

# minimum sales per resturant 
min_sales=np.min(sales_data[:,1:],axis=1)
# print("minimum sales per resturant::",min_sales)

# maximum sales per year
max_sales=np.max(sales_data[1],axis=0)
print("maximum sales::",max_sales)

# sum
total_value=np.sum(sales_data[:3,1:],axis=0)
print("this is total value::",total_value)
print()

# average sales per year
avg_sales=np.average(sales_data[:,1:],axis=0)
print("average sales per year::",avg_sales)
# average sales per resturant
avg_sales=np.average(sales_data[:,1:],axis=1)
print("average sales per resturant::\n",avg_sales)

```

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)
print("this is cumsum sales::\n",cumsum)
```

- How To Draw Graph using matplotlib (overview) :-

```python
cumsum=np.cumsum(sales_data[:,1:],axis=1)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
# 
plt.grid(True)
plt.show()
```
